### Data Pre-Processing

```python
# read data

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DataSet/Salary_Data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# check NA
data.isnull().sum()
```

```python
# check duplicate
data.duplicated().sum()
```

```python
plt.scatter(data['YearsExperience'], data['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
# split data
x = data['YearsExperience'].values.reshape((-1, 1)) # convert dataframe to array
y = data['Salary'].values.reshape((-1, 1))
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Modeling

```python
# linear regression
from sklearn.linear_model import LinearRegression

model = LinearRegression() # create object
model.fit(x_train, y_train)
```

### Evaluation

```python
# best evaluation for regression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE: ', mse)
print('MAE: ', mae)
print('R2: ', r2)
```
